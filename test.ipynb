```python
import numpy as np
from scipy.stats import norm

import matplotlib.pyplot as plt

def generate_data(shape=(100,)):
    return np.random.normal(size=shape) + 5

def posterior(mu, sigma, y):
    likel = norm.logpdf(y, mu, sigma).mean()
    prior_mu = norm.logpdf(mu, 0, 10)
    prior_sigma = norm.logpdf(sigma, 1, 0.5)
    return likel + prior_mu + prior_sigma

D = generate_data()
vals = np.zeros(shape=(100,100))
x = np.linspace(4, 6, 100)
y = np.linspace(0.7, 1.3, 100)
for i in range(len(x)):
    for j in range(len(y)):
        vals[i, j] = posterior(x[i], y[j], D)



```

```python
plt.plot(x, np.mean(vals, axis=1))
plt.show()
plt.plot(y, np.mean(vals, axis=0))
plt.show()

X, Y = np.meshgrid(x, y)
plt.contourf(X, Y, vals, levels=20)
plt.show()
```
